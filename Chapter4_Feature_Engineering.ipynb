```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

X, y = mglearn.datasets.make_wave(n_samples=100)
line = np.linspace(-3, 3, 1000, endpoint=False).reshape(-1, 1)
reg = DecisionTreeRegressor(min_samples_split=3).fit(X, y)
plt.plot(line, reg.predict(line), label='decision tree')

reg = LinearRegression().fit(X, y)
plt.plot(line, reg.predict(line),  label='linear regression')

plt.plot(X[:, 0], y, 'o', c='k')
plt.xlabel('Input feature')
plt.ylabel('Regression output')
plt.legend(loc='best')
plt.show()
```

```python
bins = np.linspace(-3, 3, 11)
print(f'bins: {bins}')
```

```python
which_bin = np.digitize(X, bins=bins)
print(f'\nData points:\n{X[:5]}')
print(f'\nBin membership for data points:\n{which_bin[:5]}')
```

```python
from sklearn.preprocessing import OneHotEncoder

# OneHotEncoderで置換する
encoder = OneHotEncoder(sparse=False)
# encoder.fitでwhich_binに現れる整数値のバリエーションを確認
encoder.fit(which_bin)
# transformでOneHotEncodingを行う
X_binned = encoder.transform(which_bin)
print(X_binned[:5])
```

```python
print(f'X_binned.shape: {X_binned.shape}')
```

```python
line_binned = encoder.transform(np.digitize(line, bins=bins))

reg = DecisionTreeRegressor(min_samples_split=3).fit(X_binned, y)
plt.plot(line, reg.predict(line_binned), label='decision tree binned')

reg = LinearRegression().fit(X_binned, y)
plt.plot(line, reg.predict(line_binned),  label='linear regression binned')

plt.plot(X[:, 0], y, 'o', c='k')
plt.xlabel('Input feature')
plt.ylabel('Regression output')
plt.legend(loc='best')
plt.show()
```

```python
X_combined = np.hstack([X, X_binned])
print(X_combined.shape)
```

```python
reg = LinearRegression().fit(X_combined, y)

line_combined = np.hstack([line, line_binned])
plt.plot(line, reg.predict(line_combined),  label='linear regression combined')


for bin in bins:
    plt.plot([bin, bin], [-3, 3], ':', c='k')
plt.plot(X[:, 0], y, 'o', c='k')
plt.xlabel('Input feature')
plt.ylabel('Regression output')
plt.legend(loc='best')
plt.show()
```

```python
X_product = np.hstack([X_binned, X * X_binned])
print(X_product.shape)
```

```python
reg = LinearRegression().fit(X_product, y)

line_product = np.hstack([line_binned, line * line_binned])
plt.plot(line, reg.predict(line_product), label='linear regression product')

for bin in bins:
    plt.plot([bin, bin], [-3, 3], ':', c='k')
    
plt.plot(X[:, 0], y, 'o', c='k')
plt.xlabel('Input feature')
plt.ylabel('Regression output')
plt.legend(loc='best')
plt.show()
```

```python
from sklearn.datasets import load_boston
from sklearn.preprocessing import MinMaxScaler

boston = load_boston()
X_train, X_test, y_train, y_test = split(boston.data, boston.target, random_state=0)

# データのスケール変換
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
# 2次までの多項式特徴量と交互作用を抽出
poly = PolynomialFeatures(degree=2).fit(X_train_scaled)
X_train_poly = poly.transform(X_train_scaled)
X_test_poly = poly.transform(X_test_scaled)
print(f'X_train.shape: {X_train.shape}')
print(f'X_train_poly.shape: {X_train_poly.shape}')
```

```python
print(f'Polynomial feature names: \n{poly.get_feature_names()}')
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train_scaled, y_train)
print('Score without interactions: {:.3f}'.format(ridge.score(X_test_scaled, y_test)))

ridge = Ridge().fit(X_train_poly, y_train)
print('Score with interactions: {:.3f}'.format(ridge.score(X_test_poly, y_test)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100).fit(X_train_scaled, y_train)
print('Scpre without interactions: {:.3f}'.format(rf.score(X_test_scaled, y_test)))

rf = RandomForestRegressor(n_estimators=100).fit(X_train_poly, y_train)
print('Scpre without interactions: {:.3f}'.format(rf.score(X_test_poly, y_test)))
```

### 単変量統計

```python
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import SelectPercentile

cancer = load_breast_cancer()

# シードを指定して乱数を決定
rng = np.random.RandomState(42)
noise = rng.normal(size=(len(cancer.data), 50))
# ノイズ特徴量をデータに加える
# 最初の30特徴量はデータセットから来たもの。続く50特徴量はノイズ
X_w_noise = np.hstack([cancer.data, noise])

X_train, X_test, y_train, y_test = split(
    X_w_noise, cancer.target, random_state=0, test_size=.5)
# f_classif(デフォルト)とSelectPercentileを使って50%の特徴量を選択
select = SelectPercentile(percentile=50)
select.fit(X_train, y_train)
# 訓練セットを変換
X_train_selected = select.transform(X_train)

print(f'X_train.shape: {X_train.shape}')
print(f'X_train_selected.shape: {X_train_selected.shape}')
```

```python
# どの特徴量が使われているか'get_support'メソッドで調べることができる
mask = select.get_support()
print(mask)
# maskを可視化する -- 黒がTrue、白がFalse
plt.matshow(mask.reshape(1, -1), cmap='gray_r')
plt.xlabel('Sample index')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

# データの変換
X_test_selected = select.transform(X_test)

lr = LogisticRegression()
lr.fit(X_train, y_train)
print('Score with all features: {:.3f}'.format(lr.score(X_test, y_test)))

lr.fit(X_train_selected, y_train)
print('Score with only selected features: {:.3f}'.format(lr.score(X_test_selected, y_test)))
```

```python
plt.figure(figsize=(10, 3))
xticks = pd.date_range(start=citibike.index.min(), end=citibike.index.max(), freq='D')
plt.xticks(xticks, xticks.strftime('%a %m-%d'), rotation=90, ha='left')
plt.plot(citibike, linewidth=1)
plt.xlabel('Date')
plt.ylabel('Rentals')
plt.show()
```

```python
# ターゲット値(レンタル数)を抽出
y = citibike.values
#　時刻を'%s'でPOSIX時刻に変換
X = citibike.index.astype('int64').values.reshape(-1, 1) // 10 ** 9
```

```python
print(f'X.shape: {X.shape}')
print(f'y.shape: {y.shape}')
print(f'y: {y[:5]}')
print(f'X: {X[:5]}')
```

```python
# 最初の184データポイントをtrainingに、残りをtestに使う
n_train = 184

# 与えられた特徴量セットで、回帰器を評価し、プロットする関数
def eval_on_features(features, target, regressor):
    # 与えられた特徴量をtrainigデータとテストデータに分割
    X_train, X_test = features[:n_train], features[:n_train]
    # ターゲットの配列も分割する
    y_train, y_test = target[:n_train], target[:n_train]
    
    regressor.fit(X_train, y_train)
    print('Test-set R^2: {:.2f}'.format(regressor.score(X_test, y_test)))
    y_pred_train = regressor.predict(X_train)
    y_pred = regressor.predict(X_test)
    
    plt.figure(figsize=(10, 3))
    plt.xticks(range(0, len(X), 8), xticks.strftime('%a %m-%d'), rotation=90, ha='left')
    
    plt.plot(range(n_train), y_train, label='train')
    plt.plot(range(n_train, len(y_test) + n_train), y_test, '-', label='test')
    
    plt.plot(range(n_train), y_pred_train, '--', label='prediction train')
    plt.plot(range(n_train, len(y_test) + n_train), y_pred, '--', label='prediction test')
    
    plt.legend(loc=(1.01, 0))
    plt.xlabel('Date')
    plt.ylabel('Rentals')
    plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=100, random_state=0)
plt.figure()
eval_on_features(X, y, regressor)
```
